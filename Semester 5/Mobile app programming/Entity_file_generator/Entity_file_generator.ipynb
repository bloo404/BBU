```python
list_items = ['id','name','breed','age','weight','owner','location', 'description']
types = [2,1,1,0,0,1,1,1] # 0 is int, 1 is string, 2 is int?, 3 is double
file = "class Entity{ \n"

for i in range(0,len(types)):
    line = ""
    if types[i]==2:
        line = 'int? '
    elif types[i]==0:
        line = 'int '
    elif types[i]==1:
        line = 'String '
    line += list_items[i] + ';'
    file+= line + '\n'

file += "Entity({"

for i in range(0, len(types)):
    if types[i]!=2:
        file += 'required '
    file += 'this.' + list_items[i]
    if i!=len(types)-1:
         file+= ',' + '\n'

file += "}); \n"

file += " factory Entity.fromJson(Map<String, dynamic> json) { \n return Entity( \n"

for i in range(0, len(types)):
    file += list_items[i] + ': json[\'' + list_items[i] + '\'] ' + 'as '
    if types[i]==2:
        file += 'int?'
    elif types[i]==1:
        file += 'String'
    elif types[i]==0:
        file += 'int'
    if i!=len(types)-1:
         file+= ',' + '\n'

file += '); \n } \n'

file += 'Map<String, dynamic> toJson() { \n return { \n '

for i in range(0, len(types)):
    file += '\'' + list_items[i] + '\':' + list_items[i]
    if i!=len(types)-1:
         file+= ',' + '\n'
            
file += '}; \n }\n Map<String, dynamic> toJsonWithoutId() { \n return { \n'

for i in range(1, len(types)):
    file += '\'' + list_items[i] + '\':' + list_items[i]
    if i!=len(types)-1:
         file+= ',' + '\n'

file += '}; \n } \n '

file += 'Entity copy({ \n'

for i in range(0, len(types)):
    line = ""
    if types[i]==2 or types[i]==0:
        line = 'int? '
    elif types[i]==1:
        line = 'String? '
    line += list_items[i]
    if i!=len(types)-1:
         line+= ','
    file+= line + '\n'
    
file += '}) => \n Entity( \n'

for i in range(0, len(types)):
    line = ""
    line += list_items[i] + ': ' + list_items[i] + ' ?? ' + 'this.' + list_items[i] 
    if i!=len(types)-1:
         line += ','
    else: line += ');'
    file+= line + '\n'

file += "@override \n String toString() { \n return \'Entity - "

for i in range(0, len(types)):
    line = ""
    line += list_items[i] + ': $' + list_items[i]
    if i!=len(types)-1:
         line += ','
    else: line += '\' ; } }'
    file+= line + ' '

    
print(file)
```

```python
# Imports

from random import randint, random, randrange, uniform
from operator import add
from math import cos, pi
from numpy import random
import numpy as np
```

```python
# Genetic algorithm with weighted n-queen problem
# 1. Initialization
def individual_matrix(chromosome_length):
    return np.random.permutation(chromosome_length)

# chromosome_length = how many queens generated
# creates a vector that the sum of represents the position on the column

def initialize_population(population_size, chromosome_length):
    return [individual_matrix(chromosome_length) for x in range(population_size)]

N = 8
population_size = N*1000 # generate a big sample because the if the sample is too small it's game over with thhe generations
chromosome_length = N
population = initialize_population(population_size, chromosome_length)
print(population)
```

```python
# 2. Fitness Evaluation (validare)

# in case it s needed to have unique solutions that do not repeat
def get_unique_arrays(arrays):
    unique_arrays = set()
    for array in arrays:
        unique_arrays.add(tuple(array))
    return [list(array) for array in unique_arrays]

# computes weight for each individual
def compute_weight(individual, weights, length):
    sum_weight = 0
    for i in range(0,length):
        sum_weight = sum_weight + weights[i][individual[i]]
        # print(weights[i][individual[i]])
    return sum_weight

# computes all weights for the whole population
def all_weights_for_population(population, weights, length):
    #for individual in population:
    #    print("------------------")
    #    print(individual)
    #    print(compute_weight(individual, weights, length))   
    return [compute_weight(individual, weights, length) for individual in population]

# computes if the queens are attacking
def function_fitness(individual, length):
    t1 = 0
    t2 = 0
    f1 = []
    f2 = []
    for i in range(0,length):
        f1.append(individual[i] - i)
        f2.append((1 + length) - individual[i] - i)
    f1.sort()
    f2.sort()
    for i in range(1,length):
        if f1[i] == f1[i-1]:
            t1 = t1 + 1
        if f2[i] == f2[i-1]:
            t2 = t2 + 1
    fitness_value = t1 + t2
    return fitness_value

# Evaluates fitness (computes fitness scores with weights inside)
def evaluate_fitness(population, weights, length):
    fitness_scores = []
    for individual in population:
        if function_fitness(individual, length) == 0:
            fitness_scores.append(compute_weight(individual, weights, length)) # adds the weights to the fitness scores list
        else:
            fitness_scores.append(compute_weight(individual,weights,length) - len(individual) * function_fitness(individual,length))
    return fitness_scores

# to generate the matrix
def randomize_matrix(rows, cols, min_val, max_val):
    matrix = [[random.randint(min_val, max_val) for _ in range(cols)] for _ in range(rows)]
    return matrix

rows = 8
cols = 8
min_val = 1
max_val = 10
weights = randomize_matrix(rows, cols, min_val, max_val)
print(weights)

# fitness scores have weights now and the population is renewed
# the villages with the queens that attack each other died
fitness_scores = evaluate_fitness(population, weights, N)

# population = get_unique_arrays(population)

# Compute all weights for them (Initially was thought for this to be separate)
all_weights = all_weights_for_population(population, weights, N)

print("---------------------------")
print("FITNESS SCORES FOR EVERY INDIVIDUAL IN POPULATION: \n =0 means no queens on diagonal, \n >0 means more queens on a diagonal")
print(fitness_scores)
print("---------------------------")
print("Sum of weights for every individual in the population: ")
print("---------------")
print("All weights:")
print(all_weights)
print("---------------------------")
```

```python
# 3. Selection

# this can be considered a tournament algorithm where we choose 
# the best two parents amongst all

def select_subsequence(population, tournament_size, fitness_scores, num_selections):
    selected = []
    pos = []
    population_size = len(population)
    
    for _ in range(int(num_selections)-1):
        tournament = random.choice(range(population_size), int(tournament_size))
        winner = None
        max_fitness = float('-inf')
        for participant in tournament:
            fitness = fitness_scores[participant]
            if fitness > max_fitness:
                max_fitness = fitness
                winner = participant
        pos.append(winner)
        selected.append(population[winner])
    return selected, pos

def select_parents(population, fitness_scores):
    # select two parents from the population based on the fitness - 
    # the better the fitness, the higher the chance to be selected
    # IN:  population, fitness_scores
    # OUT: selected_parents
    parents = []
    pos = []
    
    if fitness_scores[0] > fitness_scores[1]:
        parents.append(population[0])
        parents.append(population[1])
        maxi1 = fitness_scores[0]
        maxi2 = fitness_scores[1]
        pos.append(0)
        pos.append(1)
    else:
        parents.append(population[1])
        parents.append(population[0])
        maxi1 = fitness_scores[1]
        maxi2 = fitness_scores[0]        
        pos.append(1)
        pos.append(0)
        
    for i in range(2, len(fitness_scores)):
        if fitness_scores[i] >= maxi1:
            parents[1] = parents[0]
            maxi2 = maxi1
            parents[0] = population[i]
            maxi1 = fitness_scores[i]
            pos[1] = pos[0]
            pos[0] = i
        elif fitness_scores[i] > maxi2:
                parents[1] = population[i]
                maxi2 = fitness_scores[i]
                pos[1] = i
    return parents, pos

# parents, positions = select_parents(population, fitness_scores)
positions = []
parents, positions = select_subsequence(population, N*5, fitness_scores, N*5)
print(parents)
print(positions)
```

```python
# 4. Crossover

def crossover(parents):
    point1 = randint(0,len(parents[0])/2-1)
    point2 = randint(len(parents[0])/2,len(parents[0])-1)
    parent1 = parents[0]
    parent2 = parents[1]
    parent1_chromosomes = parent1[point1:point2]
    
    child = []
    count = 0
    x = 0
    
    while (count < point1):
        if parent2[x] not in parent1_chromosomes:
            child.append(parent2[x])
            count = count + 1
        x = x + 1
    for i in range(point1,point2 + 1):
        child.append(parent1[i])
        count = count + 1
    while (count < len(parent1)):
        if parent2[x] not in child:
            child.append(parent2[x])
            count = count + 1
        x = x + 1
    
    # this is kind of a mutation but not really, it's just the chromosomes dancing
    alpha = randint(0,len(parents))
    child = child[alpha:] + child[:alpha]

    return child

def create_new_population(parents):
    children = []
    if (len(parents)%2 == 1):
        parents.pop()
        
    for i in range(0,len(parents),2):
        pair = [parents[i],parents[i+1]]
        offspring = crossover(pair)
        children.append(offspring)
    return children
    
children = create_new_population(parents)
print(children)
```

```python
# 5. Mutation

def mutate(child, parents, mutation_rate):
    val = uniform(0.0, 1.0)
    if mutation_rate > val:
        point1 = randint(0,len(parents[0])/2-1)
        point2 = randint(len(parents[0])/2,len(parents[0])-1)
        child[point1],child[point2] = child[point2],child[point1]
    return child
    
mutation_rate = 0.1

def create_mutated_children(children, parents, mutation_rate):
    return [mutate(child, parents, mutation_rate) for child in children]

print("------MUTATED CHILDREN-----")
children = create_mutated_children(children, parents, mutation_rate)
print(children)
print("---------------------------")
```

```python
# 6. Fitness Evaluation for the new individuals: Evaluate the fitness of the new population.

fitness_scores_children_edition = evaluate_fitness(children, weights, N)
print(fitness_scores_children_edition)
```

```python
# 7. Form the new population

def select_surviving(population,fitness_scores):
    survived = []
    for i in range(len(population)):
        if fitness_scores[i]<0:
            alpha = randint(fitness_scores[i], 0)
        else:  
            alpha = randint(0, fitness_scores[i])
        if alpha > fitness_scores[i]/2:
            survived.append(population[i])
    return survived

surviving = select_surviving(children, fitness_scores_children_edition)
print(surviving)
```

```python
# 8. Generations

generations = 100

for _ in range(generations):
        family = []
        
        # Fitness evaluation
        fitness_scores = evaluate_fitness(surviving, weights, N)
        
        # Selection
        parents, pos = select_parents(surviving, fitness_scores)
        family.append(parents[0])
        family.append(parents[1])
        
        # Crossover
        offspring = crossover(family)
        
        # Mutation
        mutated_offspring = mutate(offspring, parents, mutation_rate)
        family.append(mutated_offspring)
                
        # Calculate fitness for whole family
        fitness_family = evaluate_fitness(family, weights, N)
        
        # Replace the population with the new generation
        if(fitness_family[0] < fitness_family[1]) and (fitness_family[0] < fitness_family[2]):
            surviving[pos[0]] = mutated_offspring
            
        if(fitness_family[1] < fitness_family[0]) and (fitness_family[1] < fitness_family[2]):
            surviving[pos[1]] = mutated_offspring
print(surviving)
```

```python
# 9. Termination

def best_individual(final_population):
    pos = 0
    final_fitness_scores = evaluate_fitness(final_population, weights, N)
    for i in range(len(final_fitness_scores)):
        if final_fitness_scores[i] >= final_fitness_scores[pos]:
            pos = i
    return final_population[pos], final_fitness_scores[pos]

best_individ, best_fit = best_individual(surviving)
print("-------------------------------")
print("Best individual genes are:")
print(best_individ)
print("-------------------------------")
print(f"Weight: {best_fit:.0f}")
```

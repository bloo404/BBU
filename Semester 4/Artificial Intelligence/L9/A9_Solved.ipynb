## EXERCISE 2

**Describe the search tree for the travel salesman problem in 2 ways:**

### 1) one with nodes containing possible partial solutions 

- Starts from one point which is considered the root. The rest are considered partial tour.
- Remembers: current city, visited cities, cities to visit, distance travelled
- From each node, the children are generated to add them to the route and do this untill the route is complete
- On every node, a partial solution is calculated through the length of the tour

### 2) one with full possible solutions

- It creates a complete tour for every node in the tree as the root
- At the end we shall have a node representing the full tour where we visit the root once at the beginning and at the end
- Remembers: cities visited, current city, distance traveled

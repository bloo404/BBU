```python
# EXERCISE 1

# Implement a function f for this function and 
# represent it graphic using matplotlib.

plt.rcParams["figure.figsize"] = [7.5,3.5]
plt.rcParams["figure.autolayout"] = True

# Bukin function N6
def f(x, y):
    return 100*np.sqrt(abs(y-0.01*x**2)) + 0.01 * abs (x + 10)

x = np.linspace(-15, -5, 100)
y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=[0, 2, 5, 10, 20, 50, 100, 200], colors="purple")
ax.plot([-10], [1], 'bx', markersize=10)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('EXERCISE 1 Bukin function N6')
plt.show()
```

```python
# EXERCISE 2

# Write a Hill Climbing algorithm that finds this functions minimum.

# delta =~ learning rate

def computeMinimum(fct, x, y, fct2, x2, y2):
    if fct < fct2:
        return fct, x, y
    else:
        return fct2, x2, y2

def hillClimbing_BukinN6(f, x_range, y_range, max_iter=100000, delta=0.01): 
    # Initialize the current point
    x = random.uniform(x_range[0], x_range[1])
    y = random.uniform(y_range[0], y_range[1])
    fx = f(x, y) 
    for i in range(max_iter):
        
        # values for the neighbor dude
        val = random.uniform(-delta, delta)
        xn = x + val
        yn = y + val
        xm = x - val
        ym = y - val
        
        # verify if we are within range
        if xn < x_range[0] or xn > x_range[1] or yn < y_range[0] or yn > y_range[1]:
            continue
            
        if xm < x_range[0] or xm > x_range[1] or ym < y_range[0] or ym > y_range[1]:
            continue
        
        # Computing the neighbors :P
        fxn1 = f(xn, yn)
        fxn2 = f(x, yn)
        fxn3 = f(xn, y)
        fxn4 = f(xm, ym)
        fxn5 = f(x, ym)
        fxn6 = f(xm, y)
        fxn7 = f(xn, ym)
        fxn8 = f(xm,yn)

        # Extracting the minimum and the values used for that value
        xf = 0
        yf = 0
        
        # Compute the results
        fxn, xf, yf = computeMinimum(fxn1, xn, yn, fxn2, x, yn)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn3, xn, y)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn4, xm, ym)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn5, x, ym)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn6, xm, y)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn7, xn, ym)
        fxn, xf, yf = computeMinimum(fxn, xf, yf, fxn8, xm, yn)
        
        if fxn < fx:
            x = xf
            y = yf
            fx = fxn
        if fx < 1e-6:
            break
            
    return x, y

x_range = (-15, -5)
y_range = (-3, 3)

min_x, min_y = hillClimbing_BukinN6(f, x_range, y_range)

print(f"(x,y): ({min_x}, {min_y}) \nf(x,y): {f(min_x, min_y)}")

```

```python
# EXERCISE 3

# Implement a hill climbing algorithm for the knapsack problem.

profit = [60, 100, 120, 100, 60]
weight = [10, 20, 30, 20, 30]
hasBeen = [0, 0, 0, 0, 0]
W = 50
n = len(profit)

items = []

def computeMinimum2(x1,y1, x2, y2,W,y):
    if x1 > x2 and y +  y1 <= W:
        return x1, y1
    else:
        return x2, y2

def hillClimbing_knapsack(max_iter=100000, delta = 1): 
    i = random.randint(0, n-1)
    
    # initialize the current point
    x = profit[i]
    y = weight[i]
    hasBeen[i] = 1
    fx = x
    items.append((x,y))
    for j in range(max_iter):
        
        # neighbor
        delta = random.randint(1,10)
        val = random.randint(-(10 % delta),(10 % delta))
        
        newVal = i + val
        newVal2 = i - val

        newVal = newVal % n
        newVal2 = newVal2 % n
        
        xn1 = profit[newVal]
        yn1 = weight[newVal]
        fxn1 = xn1
        
        xn2 = profit[newVal2]
        yn2 = weight[newVal2]
        fxn2 = xn2
        
        xn, yn = computeMinimum2(xn1,yn1,xn2,yn2,W,y)
        fxn = xn
        
        if fxn > fx and y + yn<=W:
            x = xn
            y = y + yn
            fx = fxn
            items.append((xn,yn))
        if y == W:
            break
    return y

final_weight = hillClimbing_knapsack()

pro = 0   
for item in items:
    pro = pro + item[0]

print(f"Final weight: {final_weight}")
print(f"Final profit: {pro}")
print("Items: ", items)
```

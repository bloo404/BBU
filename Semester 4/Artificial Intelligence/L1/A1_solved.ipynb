```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.lines import Line2D
from numpy import sin,cos,pi,linspace
import math

theta = 2 * np.pi
nr = 100
r = 1

def draw_chord(A,B,my_colour):
    x1, y1 = A
    x2, y2 = B
    plt.plot([x1,x2],[y1,y2],color = my_colour)
    
def generate():
    plt.plot(0,0, color = 'black')
    
    #draw a circle
    angles = linspace(0 * pi, 2 * pi, 100)
    r = 1

    #draw the equilateral triangle inscribed in the circle
    theta = np.linspace(0, 2*np.pi, 100)
    x1 = np.cos(theta) 
    y1 = np.sin(theta) 

    plt.plot(x1, y1)

    x = np.array([0,120,240,0])
    y = np.array([0,120,240,0])
    x = np.cos( x * np.pi / 180 )
    y = np.sin( y * np.pi / 180 )

    plt.plot( x, y, color='black' )

# workaround because jupyter ate something funny today
def generate2():
    plt.plot(0,0, color = 'black')
    
    #draw a circle
    angles = linspace(0 * pi, 2 * pi, 100)
    r = 1

    #draw the equilateral triangle inscribed in the circle
    theta = np.linspace(0, 2*np.pi, 100)
    x1 = np.cos(theta) 
    y1 = np.sin(theta) 

    plt.plot(x1, y1)
    plt.plot(x1/2,y1/2)

    x = np.array([0,120,240,0])
    y = np.array([0,120,240,0])
    x = np.cos( x * np.pi / 180 )
    y = np.sin( y * np.pi / 180 )

    plt.plot( x, y, color='black' )
    
def distance(A,B):
    return math.sqrt((B[0]-A[0])**2 + (B[1]-A[1])**2)
    
def bertrand_1():
    generate()
    
    favorable_cases = 0
    for i in range(100):
        theta = np.random.random() * 2 * pi
        x, y = np.cos(theta), np.sin(theta)
        B = x, y
        plt.plot(x,y, color='blue', marker = 'X')
        
        theta1 = np.random.random() * 2 * pi
        x1, y1 = np.cos(theta1), np.sin(theta1)
        A = x1, y1
        if distance(A,B) <= math.sqrt(3):
            draw_chord(A,B,'purple')
        else:
            draw_chord(A,B,'blue')
            favorable_cases = favorable_cases + 1

    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.gca().set_aspect('equal')
    
    probability = favorable_cases/100
    expected_value = 1/3
    
    print("Bertrand 1: \nExpected value:" + str(expected_value) + " \nOur value: " + str(probability))
    
    error = abs(1/3 - probability)
    
    print("Error: " + str(error))
    
    plt.show()
    
def bertrand_2():
    generate()
    
    favorable_cases = 0
    
    # random point on circle
    theta = np.random.random() * 2 * pi
    x, y = r * np.cos(theta), r * np.sin(theta)
    B = x, y
    plt.plot(x,y, color='blue', marker = 'X')
    
    # radius
    draw_chord((0,0), B, 'black') 
    
    m_x, m_y = r / 2 * np.cos(theta), r / 2 * np.sin(theta)
    M = m_x, m_y
    
    # angle perpendicular on the radius
    gamma = abs(theta - np.pi/2)  
    
    for i in range(100):
        # perpendiculars on radius
    
        # generate a random point on the radius
        m = np.random.random() * r
        xr, yr = m * np.cos(theta), m * np.sin(theta)
        
        # plot the chord with midpoint on the radius
        length = math.sqrt(pow(r, 2) - pow(math.sqrt(xr**2 + yr**2), 2))
        x1, y1 = xr + length * np.cos(gamma), yr + length * np.sin(gamma)
        x2, y2 = xr - length * np.cos(gamma), yr - length * np.sin(gamma)
                
        C = x1,y1
        D = x2,y2
    
        # heeee heeee when in cadran I
        # DECI SA STITI CA FORMULA ASTA E CONCEPUTA DE MINE (Bulat Jaclina-Iana) SI (Badea Dan-Nicolai) SI AM STAT 3 ORE 
        # SA O GASIM PENTRU CA IN CAZUL DE MAI SUS PENTRU FORMULA ASTA
        # DESENUL SE LASA PE SCENA CA SI MICHAEL JACKSON
        if xr>0 and yr>0:
            x1, y1 = xr + length * np.cos(theta + np.pi/2), yr + length * np.sin(theta + np.pi/2)
            x2, y2 = xr - length * np.cos(theta + np.pi/2), yr - length * np.sin(theta + np.pi/2)
            C = x1,y1
            D = x2,y2

            
        if distance(C,D) > math.sqrt(3):
            favorable_cases = favorable_cases + 1
            draw_chord(C,D,'purple')
        else:
            draw_chord(C,D,'yellow')
            
            
    plt.plot(m_x, m_y, color='red', marker = 'X')

    
    # we want a round circle, not an ellipse (visually)
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.gca().set_aspect('equal')
    
    probability = favorable_cases/100
    
    expected_value = 1/2
    
    print("Bertrand 2: \nExpected value:" + str(expected_value) + " \nOur value: " + str(probability))
    
    error = abs(1/2 - probability)
    
    print("Error: " + str(error))
    plt.show()
    
def bertrand_3():
    generate2()
    
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.gca().set_aspect('equal')    
    
    favorable_cases = 0
    
    # midpoints
    for i in range(100):
        rho = np.random.random() * 2 * pi
        m = np.random.random()
        x, y = m * np.cos(rho), m * np.sin(rho)
        plt.plot(x, y, color='red', marker = 'X')
        
        #find the chord whose midpoint is the point calculated earlier
        slope = -x/y

        h = r-(math.sqrt(x**2 + y**2))
        c = math.sqrt((r-h/2)*(8*h))
        k = math.sqrt((c**2)/(4*slope**2+4))
        
        x1 = x-k
        x2 = x+k
        y1 = y-k*slope
        y2 = y+k*slope
        V = (x1,y1)
        W = (x2,y2)
        
        if(distance(V,W) > math.sqrt(3)):
            favorable_cases = favorable_cases + 1
            draw_chord(V,W,'green')
        else:
            draw_chord(V,W,'yellow')

    
    probability = favorable_cases/100
    
    expected_value = 1/4
    
    print("Bertrand 3: \nExpected value:" + str(expected_value) + " \nOur value: " + str(probability))
     
    error = abs(1/4 - probability)
    
    print("Error: " + str(error))

    
    plt.show()
    
    
def all_Bertrands():
    bertrand_1()
    bertrand_2()
    bertrand_3()
    
all_Bertrands()
```
